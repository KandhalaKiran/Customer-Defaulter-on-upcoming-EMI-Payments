```python
#importing train dataset
import pandas as pd
df=pd.read_csv("Train.csv")
df
```

```python
N=df.fillna(0)
N
```

```python
N=N.drop(labels=['Col702','Col733','Col742','Col754','Col747','Col836','Col763','Col791','Col813','Col822','Col831','Col843','Col852'], axis=1)
N
```

```python
x_train=N.iloc[:,2:].values
x_train
```

```python
y_train=N.iloc[:,1].values
y_train
```

```python
#importing test dataset
import pandas as pd
df1=pd.read_csv("Test.csv")
df1
```

```python
M=df1.fillna(0)
M
```

```python
M=M.drop(labels=['Col702','Col733','Col742','Col747','Col836','Col754','Col763','Col791','Col813','Col822','Col831','Col843','Col852'], axis=1)
M
```

```python
x_test=M.iloc[:,1:].values
x_test
```

```python
#LinearRegression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred   #predicted output(target variable Col2)
```
